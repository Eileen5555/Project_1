```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd

base_url = "https://www.zamnesia.com/35-cannabis-seeds/295-feminized-cannabis-seeds"
seed_names = []
seed_prices = []

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive"
}

# Use a session to handle cookies
session = requests.Session()

for page_number in range(1, 64):
    # Construct the URL for the current page
    url = f"{base_url}?p={page_number}&orderby=position&orderway=asc"

    response = session.get(url, headers=headers)

    if response.status_code == 200:
        # Get the HTML content as a string
        html_content = response.text

        # Create a BeautifulSoup object
        soup = BeautifulSoup(html_content, 'html.parser')

        # Find all parent containers with class 'right_side clearfix col-lg-9 col-md-8 col-sm-12 col-xs-12'
        parent_containers = soup.find_all('div', class_='right_side clearfix col-lg-9 col-md-8 col-sm-12 col-xs-12')

        for container in parent_containers:
            # Extract seed name
            name_element = container.find('a', class_='product_link')
            name = name_element.get('title').strip() if name_element else ''
            seed_names.append(name)

            # Extract seed price
            price_span = container.find('span', class_='price')
            if price_span:
                integer_part = price_span.find('span', class_='integer_part_of_price').text.strip().replace('.', '').replace(',', '') if price_span else None
                fractional_part = price_span.find('span', class_='fractial_part_of_price').text.strip() if price_span else None
                price_str = f"{integer_part}.{fractional_part}" if integer_part and fractional_part else ''
                price_str = price_str.replace('..', '.')
                # Convert price string to float
                try:
                    price = float(price_str.replace(',', '.'))
                except ValueError:
                    price = None  # Handle non-numeric prices
                
                seed_prices.append(price)

            else:
                seed_prices.append(None)

    else:
        print(f"Failed to retrieve the webpage for page {page_number}. Status code: {response.status_code}")

# Create a DataFrame
seed_df = pd.DataFrame({'Seed Name': seed_names, 'Price (€)': seed_prices})

# Display the DataFrame
display(seed_df)
```

```python
null_counts = seed_df.isnull().sum()
print("Null values in all columns:")
print(null_counts)
```

```python
# Check for duplicate rows
duplicate_rows = seed_df[seed_df.duplicated()]

# Display duplicate rows
print("Duplicate Rows:")
display(duplicate_rows) 
```

```python
seed_df_cleaned = seed_df.copy()
seed_df_cleaned = seed_df_cleaned.drop_duplicates()
seed_df_cleaned.dropna
display(seed_df_cleaned)
```

```python
seed_df_cleaned.columns = [col.strip().lower().replace(" ", "_") for col in seed_df.columns]
seed_df_cleaned.rename(columns={'seed_name': 'name'}, inplace=True)
display(seed_df_cleaned)
```

```python
def clean_name(name):
    # Remove everything after "(" and " - " 
    name = name.split("(")[0].split(" - ")[0]
    name = name.replace(" auto", "")
    # Lowercase and strip the string
    return name.lower().strip()

# Apply the custom function to the "name" column
seed_df_cleaned["name"] = seed_df_cleaned["name"].apply(clean_name)
display(seed_df_cleaned)
```

```python
#########JOIN#######
#set name as index
df_cleaned.set_index('name', inplace=True)
seed_df_cleaned.set_index('name', inplace=True)

display(df_cleaned)
display(seed_df_cleaned)
```

```python
result = df_cleaned.join(seed_df_cleaned, how='inner')
display(result)
```
